### support_df에 대한 수정

```python
import os
par_dir = os.path.abspath(os.path.join(os.getcwd(),os.pardir))
os.chdir(par_dir)
```

```python
import pandas as pd
df = pd.read_csv('211210-DRM-total.csv')
```

```python
import numpy as np
import pandas as pd

metal_df   = df.iloc[:, 2:25].copy()
support_df = df.iloc[:, 25:70].copy()

support_sum = support_df.sum(axis=1)

# 0 아닌 행만 마스크
mask = support_sum > 0

# 복사
metal_norm   = metal_df.copy()
support_norm = support_df.copy()

# 0 아닌 행만 정규화
metal_norm.loc[mask]   = metal_df.loc[mask].div(1, axis=0)
support_norm.loc[mask] = support_df.loc[mask].div(support_sum.loc[mask], axis=0)

# 확인
print("전체 행 수:", len(df))
print("support 합 분포:")
print(support_norm.sum(axis=1).round(6).value_counts().head())
```
